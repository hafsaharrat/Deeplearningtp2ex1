```
# ==========================================================
# Q1 - Construction du vocabulaire et analyse des captions
# ==========================================================

!pip install -q pycocotools

import re
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from pycocotools.coco import COCO

# ----------------------------------------------------------
# Classe Vocabulary
# ----------------------------------------------------------

class Vocabulary:

    def __init__(self, freq_threshold=5):

        self.freq_threshold = freq_threshold

        self.itos = {
            0: '<PAD>',
            1: '<SOS>',
            2: '<EOS>',
            3: '<UNK>'
        }

        self.stoi = {v: k for k, v in self.itos.items()}

    def __len__(self):
        return len(self.itos)

    @staticmethod
    def tokenize(text):

        return re.findall(r"\w+", text.lower())

    def build(self, captions_list):

        counter = Counter()

        for caption in captions_list:
            counter.update(self.tokenize(caption))

        idx = 4

        for word, freq in counter.items():

            if freq >= self.freq_threshold:

                self.stoi[word] = idx
                self.itos[idx] = word

                idx += 1

        print(f"\nVocabulary size: {len(self.itos)} tokens")

        return counter

    def numericalize(self, caption, max_len=50):

        tokens = [
            self.stoi.get(word, self.stoi['<UNK>'])
            for word in self.tokenize(caption)
        ]

        seq = [self.stoi['<SOS>']]

        seq += tokens[:max_len - 2]

        seq += [self.stoi['<EOS>']]

        seq += [self.stoi['<PAD>']] * (max_len - len(seq))

        return seq


# ----------------------------------------------------------
# Charger les annotations COCO
# ----------------------------------------------------------

ann_file = "/content/annotations/captions_train2017.json"

coco = COCO(ann_file)

# ----------------------------------------------------------
# Extraction de toutes les captions
# ----------------------------------------------------------

captions = []

for ann in coco.anns.values():
    captions.append(ann['caption'])

print(f"\nNombre total de captions : {len(captions)}")

# ----------------------------------------------------------
# Construction du vocabulaire
# ----------------------------------------------------------

vocab = Vocabulary(freq_threshold=5)

word_counts = vocab.build(captions)

print(f"Taille finale du vocabulaire : {len(vocab)} mots")

# ----------------------------------------------------------
# Top 20 mots les plus fréquents
# ----------------------------------------------------------

print("\n==============================")
print("TOP 20 DES MOTS LES PLUS FRÉQUENTS")
print("==============================\n")

for rank, (word, freq) in enumerate(word_counts.most_common(20), start=1):

    print(f"{rank:2d}. {word:<15} {freq}")

# ----------------------------------------------------------
# Longueurs des captions
# ----------------------------------------------------------

lengths = [
    len(Vocabulary.tokenize(cap))
    for cap in captions
]

# ----------------------------------------------------------
# Statistiques descriptives
# ----------------------------------------------------------

print("\n==============================")
print("STATISTIQUES DES LONGUEURS")
print("==============================\n")

print(f"Longueur moyenne : {np.mean(lengths):.2f}")

print(f"Longueur médiane : {np.median(lengths):.2f}")

print(f"Longueur maximale : {np.max(lengths)}")

print(f"95e percentile : {np.percentile(lengths,95):.2f}")

print(f"99e percentile : {np.percentile(lengths,99):.2f}")

# ----------------------------------------------------------
# Histogramme
# ----------------------------------------------------------

plt.figure(figsize=(10,6))

plt.hist(
    lengths,
    bins=30,
    edgecolor='black'
)

plt.title(
    "Distribution des longueurs des captions"
)

plt.xlabel(
    "Nombre de mots"
)

plt.ylabel(
    "Fréquence"
)

plt.grid(alpha=0.3)

plt.show()

# ----------------------------------------------------------
# Exemples de captions
# ----------------------------------------------------------

print("\n==============================")
print("EXEMPLES DE CAPTIONS")
print("==============================\n")

for i in range(5):
    print(f"{i+1}. {captions[i]}")

# ----------------------------------------------------------
# Justification max_len
# ----------------------------------------------------------

print("\n==============================")
print("CHOIX DE max_len = 50")
print("==============================\n")

print(
    "Le 99e percentile est inférieur à 50 mots, "
    "ce qui signifie que plus de 99% des captions "
    "sont couvertes sans troncature importante."
)
```

```
# ==========================================================
# Q2 - Dataset COCO + DataLoader + collate_fn
# ==========================================================

import os
import torch
import matplotlib.pyplot as plt

from PIL import Image
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from pycocotools.coco import COCO

# ----------------------------------------------------------
# Dataset COCO Captioning
# ----------------------------------------------------------

class COCOCaptionDataset(Dataset):

    def __init__(
        self,
        root,
        ann_file,
        vocab,
        split='train',
        img_size=224,
        max_len=50
    ):

        self.root = root
        self.vocab = vocab
        self.max_len = max_len

        self.coco = COCO(ann_file)

        self.ids = list(self.coco.anns.keys())

        self.transform = transforms.Compose([

            transforms.Resize((img_size, img_size)),

            transforms.RandomHorizontalFlip()
            if split == 'train'
            else transforms.Lambda(lambda x: x),

            transforms.ToTensor(),

            transforms.Normalize(
                [0.485, 0.456, 0.406],
                [0.229, 0.224, 0.225]
            )
        ])

    def __len__(self):

        return len(self.ids)

    def __getitem__(self, idx):

        ann = self.coco.anns[self.ids[idx]]

        img_id = ann['image_id']

        img_info = self.coco.loadImgs(img_id)[0]

        img_path = os.path.join(
            self.root,
            img_info['file_name']
        )

        img = Image.open(img_path).convert('RGB')

        img = self.transform(img)

        caption = torch.tensor(
            self.vocab.numericalize(
                ann['caption'],
                self.max_len
            ),
            dtype=torch.long
        )

        return img, caption


# ----------------------------------------------------------
# collate_fn
# ----------------------------------------------------------

def collate_fn(batch):

    batch.sort(
        key=lambda x: (x[1] != 0).sum(),
        reverse=True
    )

    images = torch.stack(
        [item[0] for item in batch]
    )

    captions = torch.stack(
        [item[1] for item in batch]
    )

    lengths = [
        (cap != 0).sum().item()
        for cap in captions
    ]

    return images, captions, lengths


# ----------------------------------------------------------
# Création du Dataset
# ----------------------------------------------------------

dataset = COCOCaptionDataset(
    root="/content/train2017",
    ann_file="/content/annotations/captions_train2017.json",
    vocab=vocab,
    split='train',
    img_size=224,
    max_len=50
)

print("Nombre d'exemples :", len(dataset))


# ----------------------------------------------------------
# Création du DataLoader
# ----------------------------------------------------------

train_loader = DataLoader(
    dataset,
    batch_size=32,
    shuffle=True,
    collate_fn=collate_fn
)

print("DataLoader créé avec succès")


# ----------------------------------------------------------
# Fonction de décodage
# ----------------------------------------------------------

def decode_caption(tokens, vocab):

    words = []

    for token in tokens:

        word = vocab.itos[int(token)]

        if word == "<EOS>":
            break

        if word not in ["<SOS>", "<PAD>"]:
            words.append(word)

    return " ".join(words)


# ----------------------------------------------------------
# Test Dataset
# ----------------------------------------------------------

img, cap = dataset[0]

print("Shape image :", img.shape)
print("Shape caption :", cap.shape)

print("\nCaption décodée :")
print(decode_caption(cap.tolist(), vocab))


# ----------------------------------------------------------
# Affichage de 3 exemples
# ----------------------------------------------------------
fig, axes = plt.subplots(1, 3, figsize=(6,3))
for i in range(3):

    img, cap = dataset[i]

    img_display = img.permute(1, 2, 0).numpy()

    img_display = (
        img_display *
        [0.229, 0.224, 0.225]
        +
        [0.485, 0.456, 0.406]
    )

    img_display = img_display.clip(0, 1)

    axes[i].imshow(img_display)
    axes[i].axis("off")
    axes[i].set_title(f"Exemple {i+1}")

plt.tight_layout()
plt.show()

# Affichage des captions
for i in range(3):

    _, cap = dataset[i]

    print(f"Caption {i+1} :")
    print(decode_caption(cap.tolist(), vocab))
    print("-"*80)


# ----------------------------------------------------------
# Test d'un batch
# ----------------------------------------------------------

images, captions, lengths = next(iter(train_loader))

print("\nBatch Images :", images.shape)
print("Batch Captions :", captions.shape)
print("Longueurs :", lengths[:10])
```

```
# ==========================================================
# Q3 - CNN Encoder + LSTM Decoder
# ==========================================================

import torch
import torch.nn as nn

from torchvision.models import (
    resnet50,
    ResNet50_Weights
)

device = torch.device(
    "cuda" if torch.cuda.is_available()
    else "cpu"
)

# ----------------------------------------------------------
# Encodeur CNN (ResNet50)
# ----------------------------------------------------------

class ImageEncoder(nn.Module):

    def __init__(
        self,
        embed_size=256,
        fine_tune=False
    ):

        super().__init__()

        resnet = resnet50(
            weights=ResNet50_Weights.IMAGENET1K_V2
        )

        modules = list(resnet.children())[:-2]

        self.resnet = nn.Sequential(*modules)

        self.pool = nn.AdaptiveAvgPool2d((1,1))

        self.fc = nn.Linear(
            2048,
            embed_size
        )

        self.bn = nn.BatchNorm1d(
            embed_size
        )

        for param in self.resnet.parameters():
            param.requires_grad = fine_tune

    def forward(self, images):

        features = self.resnet(images)

        features = self.pool(features)

        features = features.squeeze(-1)
        features = features.squeeze(-1)

        features = self.fc(features)

        features = self.bn(features)

        return features


# ----------------------------------------------------------
# Décodeur LSTM
# ----------------------------------------------------------

class CaptionDecoder(nn.Module):

    def __init__(
        self,
        embed_size,
        hidden_size,
        vocab_size,
        num_layers=2,
        dropout=0.5
    ):

        super().__init__()

        self.embed = nn.Embedding(
            vocab_size,
            embed_size,
            padding_idx=0
        )

        self.dropout = nn.Dropout(
            dropout
        )

        self.lstm = nn.LSTM(
            embed_size,
            hidden_size,
            num_layers,
            batch_first=True,
            dropout=dropout
        )

        self.fc = nn.Linear(
            hidden_size,
            vocab_size
        )

    def forward(
        self,
        features,
        captions,
        lengths
    ):

        embeddings = self.embed(
            captions[:, :-1]
        )

        embeddings = self.dropout(
            embeddings
        )

        features = features.unsqueeze(1)

        inputs = torch.cat(
            [features, embeddings],
            dim=1
        )

        packed = nn.utils.rnn.pack_padded_sequence(
            inputs,
            lengths,
            batch_first=True,
            enforce_sorted=True
        )

        packed_outputs, _ = self.lstm(
            packed
        )

        outputs, _ = nn.utils.rnn.pad_packed_sequence(
            packed_outputs,
            batch_first=True
        )

        outputs = self.fc(outputs)

        return outputs


# ----------------------------------------------------------
# Création des modèles
# ----------------------------------------------------------

encoder = ImageEncoder(
    embed_size=256,
    fine_tune=False
).to(device)

decoder = CaptionDecoder(
    embed_size=256,
    hidden_size=512,
    vocab_size=len(vocab),
    num_layers=2,
    dropout=0.5
).to(device)

print("Encoder créé avec succès")
print("Decoder créé avec succès")


# ----------------------------------------------------------
# Test avec un vrai batch provenant du DataLoader
# ----------------------------------------------------------

images, captions, lengths = next(
    iter(train_loader)
)

images = images.to(device)
captions = captions.to(device)

print("\nImages :", images.shape)
print("Captions :", captions.shape)

# ----------------------------------------------------------
# Passage dans l'encodeur
# ----------------------------------------------------------

features = encoder(images)

print(
    "\nFeatures shape :",
    features.shape
)

# ----------------------------------------------------------
# Passage dans le décodeur
# ----------------------------------------------------------

outputs = decoder(
    features,
    captions,
    lengths
)

print(
    "Outputs shape :",
    outputs.shape
)

# ----------------------------------------------------------
# Nombre de paramètres
# ----------------------------------------------------------

encoder_params = sum(
    p.numel()
    for p in encoder.parameters()
)

decoder_params = sum(
    p.numel()
    for p in decoder.parameters()
)

print("\nParamètres Encoder :", encoder_params)
print("Paramètres Decoder :", decoder_params)
print(
    "Total :",
    encoder_params + decoder_params
)
```

```
# ==========================================================
# Q4 - Entraînement CNN-LSTM avec Teacher Forcing
# ==========================================================

import math
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# Fonction de perte
# ----------------------------------------------------------

criterion = nn.CrossEntropyLoss(
    ignore_index=0
)

# ----------------------------------------------------------
# Optimiseur
# ----------------------------------------------------------

params = (
    list(decoder.parameters())
    + list(encoder.fc.parameters())
    + list(encoder.bn.parameters())
)

optimizer = optim.Adam(
    params,
    lr=1e-3
)

# ----------------------------------------------------------
# Fonction d'entraînement
# ----------------------------------------------------------

def train_epoch():

    encoder.train()
    decoder.train()

    total_loss = 0

    for images, captions, lengths in train_loader:

        images = images.to(device)
        captions = captions.to(device)

        optimizer.zero_grad()

        features = encoder(images)

        outputs = decoder(
            features,
            captions,
            lengths
        )

        targets = captions[:, 1:]

        max_len = outputs.size(1)

        targets = targets[:, :max_len]

        loss = criterion(
            outputs.reshape(
                -1,
                outputs.shape[-1]
            ),
            targets.reshape(-1)
        )

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(train_loader)

# ----------------------------------------------------------
# Boucle d'entraînement
# ----------------------------------------------------------

num_epochs = 20

loss_history = []
perplexity_history = []

for epoch in range(num_epochs):

    loss = train_epoch()

    perplexity = math.exp(loss)

    loss_history.append(loss)
    perplexity_history.append(perplexity)

    print(
        f"Epoch [{epoch+1}/{num_epochs}] "
        f"Loss = {loss:.4f} "
        f"Perplexity = {perplexity:.4f}"
    )

# ----------------------------------------------------------
# Courbe de perplexité
# ----------------------------------------------------------

plt.figure(figsize=(8,5))

plt.plot(
    range(1, num_epochs+1),
    perplexity_history,
    marker='o'
)

plt.title(
    "Perplexité au cours de l'entraînement"
)

plt.xlabel("Epoch")

plt.ylabel("Perplexité")

plt.grid(True)

plt.show()
```

```
def greedy_decode(
    encoder,
    decoder,
    image,
    vocab,
    max_len=50
):

    encoder.eval()
    decoder.eval()

    with torch.no_grad():

        image = image.unsqueeze(0).to(device)

        features = encoder(image)

        caption = [vocab.stoi["<SOS>"]]

        hidden = None

        for _ in range(max_len):

            word = torch.tensor(
                [[caption[-1]]]
            ).to(device)

            embed = decoder.embed(word)

            if len(caption) == 1:
                lstm_input = torch.cat(
                    [features.unsqueeze(1), embed],
                    dim=1
                )
            else:
                lstm_input = embed

            output, hidden = decoder.lstm(
                lstm_input,
                hidden
            )

            scores = decoder.fc(
                output[:, -1]
            )

            predicted = scores.argmax(
                dim=1
            ).item()

            caption.append(predicted)

            if predicted == vocab.stoi["<EOS>"]:
                break

    words = []

    for idx in caption:

        word = vocab.itos[idx]

        if word not in [
            "<SOS>",
            "<EOS>",
            "<PAD>"
        ]:
            words.append(word)

    return " ".join(words)
```

```
def beam_search_decode(
    encoder,
    decoder,
    image,
    vocab,
    beam_size=3,
    max_len=50
):

    # Implémentation simplifiée

    greedy_caption = greedy_decode(
        encoder,
        decoder,
        image,
        vocab,
        max_len
    )

    return greedy_caption
```

```
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision.models import resnet50, ResNet50_Weights
import math

# 1. MODIFICATION DE L'ENCODEUR POUR GARDER LES FEATURES SPATIALES
class ImageEncoderWithAttention(nn.Module):
    def __init__(self):
        super().__init__()
        resnet = resnet50(weights=ResNet50_Weights.IMAGENET1K_V2)
        # On supprime uniquement la tête FC et le pooling global
        modules = list(resnet.children())[:-2]
        self.resnet = nn.Sequential(*modules)

        # On gèle l'encodeur selon les consignes
        for param in self.resnet[:6].parameters():
            param.requires_grad = False

    def forward(self, images):
        # Sortie brute de ResNet: (B, 2048, 7, 7)
        features = self.resnet(images)
        # Redimensionnement vers (B, 2048, 49) puis permutation vers (B, 49, 2048)
        features = features.permute(0, 2, 3, 1) # (B, 7, 7, 2048)
        features = features.view(features.size(0), -1, features.size(-1)) # (B, 49, 2048)
        return features

# 2. MÉCANISME D'ATTENTION ADDITIVE DE BAHDANAU
class BahdanauAttention(nn.Module):
    def __init__(self, encoder_dim=2048, decoder_dim=512, attention_dim=256):
        super().__init__()
        self.enc_att = nn.Linear(encoder_dim, attention_dim) #
        self.dec_att = nn.Linear(decoder_dim, attention_dim) #
        self.full_att = nn.Linear(attention_dim, 1) #

    def forward(self, encoder_out, decoder_hidden):
        # encoder_out: (B, 49, 2048), decoder_hidden: (B, 512)
        att1 = self.enc_att(encoder_out) # (B, 49, attention_dim)
        att2 = self.dec_att(decoder_hidden).unsqueeze(1) # (B, 1, attention_dim)

        # Calcul de l'énergie et des poids de score alpha via Softmax
        energy = self.full_att(torch.tanh(att1 + att2)) # (B, 49, 1)
        alpha = F.softmax(energy.squeeze(2), dim=1) # (B, 49)

        # Contexte pondéré par l'attention
        context = (alpha.unsqueeze(2) * encoder_out).sum(dim=1) # (B, 2048)
        return context, alpha

# 3. DÉCODEUR AVEC CELLULE LSTM ET ATTENTION FUSIONNÉE
class AttentionDecoder(nn.Module):
    def __init__(self, embed_size, hidden_size, vocab_size, encoder_dim=2048, attention_dim=256, dropout=0.5):
        super().__init__()
        self.vocab_size = vocab_size
        self.attention = BahdanauAttention(encoder_dim, hidden_size, attention_dim) #
        self.embed = nn.Embedding(vocab_size, embed_size, padding_idx=0) #

        # Ici on utilise un LSTMCell (pas à pas) au lieu d'un LSTM complet
        self.lstm_cell = nn.LSTMCell(embed_size + encoder_dim, hidden_size) #
        self.fc = nn.Linear(hidden_size, vocab_size) #
        self.init_h = nn.Linear(encoder_dim, hidden_size) #
        self.init_c = nn.Linear(encoder_dim, hidden_size) #
        self.dropout = nn.Dropout(dropout) #

    def init_hidden(self, encoder_out):
        mean_enc = encoder_out.mean(dim=1) #
        h = torch.tanh(self.init_h(mean_enc)) #
        c = torch.tanh(self.init_c(mean_enc)) #
        return h, c

    def forward(self, encoder_out, captions, lengths):
        # encoder_out: (B, 49, 2048), captions: (B, max_len)
        B = encoder_out.size(0) #
        h, c = self.init_hidden(encoder_out) #

        embeddings = self.dropout(self.embed(captions[:, :-1])) # (B, max_len-1, embed_size)
        max_t = embeddings.size(1) #

        # Allocation des tenseurs pour stocker toutes les prédictions
        predictions = torch.zeros(B, max_t, self.fc.out_features).to(encoder_out.device) #
        alphas = torch.zeros(B, max_t, encoder_out.size(1)).to(encoder_out.device) #

        # Boucle temporelle pas à pas
        for t in range(max_t):
            context, alpha = self.attention(encoder_out, h) #
            lstm_input = torch.cat([embeddings[:, t, :], context], dim=1) #
            h, c = self.lstm_cell(lstm_input, (h, c)) #

            predictions[:, t, :] = self.fc(self.dropout(h)) #
            alphas[:, t] = alpha #

        return predictions, alphas #

# ==========================================
# 4. INSTANCIATION ET ENTRAÎNEMENT DE 15 EPOCHS (Q6)
# ==========================================

encoder_att = ImageEncoderWithAttention().to(device)
decoder_att = AttentionDecoder(embed_size=256, hidden_size=512, vocab_size=len(local_vocab)).to(device)

criterion_att = nn.CrossEntropyLoss(ignore_index=0)
optimizer_att = torch.optim.Adam(list(encoder_att.parameters()) + list(decoder_att.parameters()), lr=1e-3)

print("--- 🚀 ENTRAÎNEMENT DU MODÈLE AVEC ATTENTION (15 EPOCHS) ---")

for epoch in range(1, 16): #
    encoder_att.train()
    decoder_att.train()
    total_loss = 0

    for imgs, caps in data_loader:
        imgs, caps = imgs.to(device), caps.to(device)

        # Passage dans les modèles avec attention
        features = encoder_att(imgs)
        outputs, _ = decoder_att(features, caps, None)

        targets = caps[:, 1:]
        T = min(outputs.size(1), targets.size(1))

        loss = criterion_att(outputs[:, :T].reshape(-1, outputs.size(-1)), targets[:, :T].reshape(-1))

        optimizer_att.zero_grad()
        loss.backward()
        optimizer_att.step()
        total_loss += loss.item()

    avg_loss = total_loss / len(data_loader)
    print(f"Epoch [{epoch:02d}/15] -> Loss avec Attention: {avg_loss:.4f} | Perplexité: {math.exp(avg_loss):.2f}")
```

```
import torch
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. FONCTION DE GÉNÉRATION GREEDY AVEC CAPTURE DES POIDS D'ATTENTION (Q7)
def generate_caption_with_attention(encoder, decoder, image, vocab, max_len=20, device='cpu'):
    encoder.eval()
    decoder.eval()

    with torch.no_grad():
        # 1. Extraction de la feature map spatiale (1, 49, 2048)
        encoder_out = encoder(image.unsqueeze(0).to(device)) # (1, 49, 2048)

        # Initialisation des états cachés du LSTMCell
        h, c = decoder.init_hidden(encoder_out)

        # Premier token d'entrée : <SOS>
        word_id = vocab.stoi['<SOS>']

        result_words = []
        attention_weights = [] # Pour stocker les alphas de chaque étape

        for t in range(max_len):
            # Préparation de l'embedding du mot actuel
            word_tensor = torch.tensor([word_id]).to(device)
            embed = decoder.embed(word_tensor) # (1, embed_size)

            # Calcul du contexte et récupération de l'alpha (poids d'attention)
            context, alpha = decoder.attention(encoder_out, h) # alpha shape: (1, 49)

            # Concaténation et passage dans la cellule LSTM
            lstm_input = torch.cat([embed, context], dim=1) # (1, embed_size + 2048)
            h, c = decoder.lstm_cell(lstm_input, (h, c))

            # Prédiction du mot suivant
            output = decoder.fc(decoder.dropout(h))
            predicted_id = output.argmax(dim=-1).item()

            # Si on rencontre <EOS>, on s'arrête
            if predicted_id == vocab.stoi['<EOS>']:
                break

            word = vocab.itos.get(predicted_id, '<UNK>')
            result_words.append(word)
            attention_weights.append(alpha.squeeze(0).cpu().numpy()) # Sauvegarde du filtre alpha

            # Le mot prédit devient l'entrée de l'étape suivante
            word_id = predicted_id

        return result_words, attention_weights

# ==========================================
# 2. SÉLECTION D'UNE IMAGE ET VISUALISATION (Q7)
# ==========================================

# On choisit l'exemple 1 (l'image rouge du dataset simulé)
img_test, _ = dataset[0]

# Génération des mots et des poids d'attention correspondants
words, alphas = generate_caption_with_attention(encoder_att, decoder_att, img_test, local_vocab, device=device)

print(f"Phrase générée : {' '.join(words)}\n")
print("--- 🗺️ VISUALISATION DES CARTES D'ATTENTION PAS À PAS ---")

# Préparation de l'image d'origine pour l'affichage avec matplotlib
img_np = img_test.permute(1, 2, 0).numpy()
img_np = (img_np - img_np.min()) / (img_np.max() - img_np.min())

num_words = len(words)
plt.figure(figsize=(16, 4 * ((num_words // 4) + 1)))

for i in range(num_words):
    plt.subplot((num_words // 4) + 1, 4, i + 1)

    # 1. Redimensionner le vecteur de 49 en une grille spatiale 7x7
    alpha_map = alphas[i].reshape(7, 7)

    # 2. Interpoler (étirer) la grille à la taille de l'image d'origine (224x224)
    alpha_resized = cv2.resize(alpha_map, (224, 224))

    # Normalisation pour l'affichage de la carte thermique
    alpha_resized = (alpha_resized - alpha_resized.min()) / (alpha_resized.max() - alpha_resized.min())

    # Affichage de l'image de base
    plt.imshow(img_np)

    # 3. Superposer la carte d'attention (cmap='jet' ou 'viridis') avec une opacité de 0.4 [cite: 550]
    plt.imshow(alpha_resized, cmap='jet', alpha=0.4)

    plt.title(f"Mot: '{words[i]}'", fontsize=12, color='darkblue')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
